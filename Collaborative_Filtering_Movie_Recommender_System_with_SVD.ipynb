# Merge required datasets

```python
merged_dataset= pd.merge(pd.merge(ratings,movies , how="inner", on='movieId'), links, how="inner",on='movieId')
merged_dataset["genres"] = merged_dataset["genres"].str.replace('|',' , ') # Arrange punctutıon which looking bad on Genres column
merged_dataset
```

# Creating a final refined dataset with unique user id, movie name combination and their ratings:

```python
# Dataset created from the existing merged dataset by grouping the unique user.. 
# id and movie title combination and potential the ratings by a user might to the same.. 
# movie in different  timestamps are averaged and stored in the new dataset.

refined_dataset = merged_dataset.groupby(by=['userId','title', "imdbId", "tmdbId", "genres","movieId"], as_index=False).agg({"rating":"mean"})
refined_dataset

# So that, getting mean of ratings according to per Unique User.
```

# Converting user id, title column of refined dataset to respective lists:

```python
#list of all users
unique_users = refined_dataset['userId'].unique() 
#creating a list of all movie names in it
unique_movies = refined_dataset['title'].unique()
len(unique_movies),len(unique_users)
```

```python
users_list = refined_dataset['userId'].tolist()
movie_list = refined_dataset['title'].tolist()
ratings_list = refined_dataset['rating'].tolist()

len(users_list),len(movie_list), len(ratings_list)
```

```python
# Creating a dict to map movies name to their corresponding index in the unique movie name list
movies_dict = {unique_movies[i] : i for i in range(len(unique_movies))}
print(len(movies_dict))
```

```python
# Creating Utility matrix with rows as movies, columns ad users to make the refined dataframe compatible for SVD implementations.
# Creating an empty array with (number of rows = number of movies)..
# and (number of columns = number of users) rows as movies, columns as users.
utility_matrix = np.asarray([[np.nan for j in range(len(unique_users))] for i in range(len(unique_movies))])

print("Shape of Utility matrix: ", utility_matrix.shape)

for i in range(len(ratings_list)):

## entry in users list and subtract 1 to get the index, we do the same for movies but we already defined a dictionary to get the index.
  utility_matrix[movies_dict[movie_list[i]]][users_list[i]-1] = ratings_list[i]

utility_matrix
```

```python
#  Normalize the utility matrix across movies column

mask = np.isnan(utility_matrix)
masked_arr = np.ma.masked_array(utility_matrix, mask)
temp_mask = masked_arr.T
rating_means = np.mean(temp_mask, axis=0)

filled_matrix = temp_mask.filled(rating_means)
filled_matrix = filled_matrix.T
filled_matrix = filled_matrix - rating_means.data[:,np.newaxis]
```

```python
filled_matrix = filled_matrix.T / np.sqrt(len(movies_dict)-1)
filled_matrix.shape
```

- Mean values across movies columns are extracted.
- Nan's of utility matrix are imputed with mean values extracted.
- Later the utility matrix has been normalized across movies to get all ratings to a standard/normal scale

```python
# Computing SVD (Singular Value Decomposition) of Utility Matrix
U, S, V = np.linalg.svd(filled_matrix)
```

```python
#  Creating a list of Case insensitive movie names for effective use
case_insensitive_movies_list = [i.lower() for i in unique_movies]
```

```python
#Function to calculate the cosine similarity (sorting by most similar and returning the top N)

def top_cosine_similarity(data, movie_id, top_n=10):  # Defining a function to calculate the cosine similarity on the given dataframe
    index = movie_id
    movie_row = data[index, :]
    magnitude = np.sqrt(np.einsum('ij, ij -> i', data, data  # extracting requesting number of closely matched movie indices with the help of numpy einsum which valuates..
                                                             #the Einstein summation convention on the operands.
    similarity = np.dot(movie_row, data.T) / (magnitude[index] * magnitude)
    sort_indexes = np.argsort(-similarity)
    
    return sort_indexes[:top_n]
```

#  Defining a function to get similar movies for the given movie name

```python
#k-principal components to represent movies, 
#movie_id to find recommendations, 
# top_n print n results        

def round_to_nearest_half_int(num): # This function for round to float nearest decimal
                                    # exp : 3.35 > 3.5
    return round(num * 2) / 2


def get_similar_movies(movie_name,top_n,k = 20):
    
    sliced = V.T[:, :k] # representative data
  # k = 50
  # movie_id = 1
  # top_n = 10 

    movie_id = movies_dict[movie_name]
    indexes = top_cosine_similarity(sliced, movie_id, top_n)
    print(" ")
    print("Top",top_n, "movies which are very much similar to the Movie-",movie_name, "are: ")
    print(" ")
    
    for i in indexes[0:]:
        print(unique_movies[i])
        
        genres= refined_dataset[refined_dataset["title"]== unique_movies[i]]["genres"].loc[1:].tolist()[0] # it will print genres according to unique movie name
        print(f"Genres: {genres}")
        
        average_voted= refined_dataset[refined_dataset["title"]== unique_movies[i]]["rating"].mean()
        print(f"Those who watched this movie rated average of {round_to_nearest_half_int(average_voted)} out of 5.")  # it will print  rated average of unique movie
        
        imdb_link = refined_dataset[refined_dataset["title"]== unique_movies[i]]["imdbId"].loc[1:].tolist()[0]  
        print(f"IMDB Link: http://www.imdb.com/title/tt0{imdb_link}")    # # it will print link code of IMDB of unique movie
        
        tmdb_link = refined_dataset[refined_dataset["title"]== unique_movies[i]]["tmdbId"].loc[1:].tolist()[0]
        print(f"TMDB Link: https://www.themoviedb.org/movie/{tmdb_link}") # it will print link code of TMDB of unique movie
        
        print("")
```

- The SVD decreases the dimension of the utility matrix A by extracting its latent factors. It maps each user and each item into a r-dimensional latent space. This mapping facilitates a clear representation of relationships between users and items.

# Defining a function which outputs movie names as suggestion when the user mis spells the movie name. User might have intended to type any of these movie names.

```python
# function which takes input and returns suggestions for the user
def get_possible_movies(movie):
    
    temp = ''
    possible_movies = case_insensitive_movies_list.copy()
    for i in movie :
        out = []
        temp += i
        
        for j in possible_movies:
            if temp in j:
                out.append(j)
        if len(out) == 0:
            return possible_movies
        
        out.sort()
        possible_movies = out.copy()
        
    return possible_movies
```

- This function provides user with movie name suggestions if movie name is mis-spelled or Recommends similar movies to the input movie if the movie name is valid.

```python
class invalid(Exception):
    pass

def recommender():

    
    try:

        movie_name = input("Please enter the movie name to we can recommend movie to you: ")
        movie_name_lower = movie_name.lower()
        
        if movie_name_lower not in case_insensitive_movies_list :
            raise invalid
        else :

            num_recom = int(input("Enter Number of movie recommendations needed: "))
            get_similar_movies(unique_movies[case_insensitive_movies_list.index(movie_name_lower)],num_recom)

    except invalid:
        
        possible_movies = get_possible_movies(movie_name_lower)
        
        if len(possible_movies) == len(unique_movies) :
            print("Movie name entered is does not exist in the list ")
            
        else :
            indices = [case_insensitive_movies_list.index(i) for i in possible_movies]
            print("Entered Movie name is not matching with any movie from the dataset . Please check the below suggestions :\n") 
            for i in indices[0:]: 
                print(unique_movies[i])
                
                genres= refined_dataset[refined_dataset["title"]== unique_movies[i]]["genres"].loc[1:].tolist()[0]
                print(f"Genres: {genres}")
                
                average_voted= refined_dataset[refined_dataset["title"]== unique_movies[i]]["rating"].mean()
                print(f"Those who watched this movie rated average of {round_to_nearest_half_int(average_voted)} out of 5.")
                
                imdb_link = refined_dataset[refined_dataset["title"]== unique_movies[i]]["imdbId"].loc[1:].tolist()[0]
                print(f"IMDB Link: http://www.imdb.com/title/tt0{imdb_link}")
                
                tmdb_link = refined_dataset[refined_dataset["title"]== unique_movies[i]]["tmdbId"].loc[1:].tolist()[0]
                print(f"TMDB Link: https://www.themoviedb.org/movie/{tmdb_link}")
                
                print("")
            
            print("")
            
            
            
            recommender()
            
        
```
