```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import string
import pickle
```

```python
dataset = pickle.load(open("ratings-amazon.pkl","rb" ))
```

```python
dataset
```

```python
data  = pd.DataFrame(dataset["itemId"].unique(), columns = {"itemId"})
```

```python
data= pd.merge(data, dataset[["itemId", "title", "brand", "category","main_cat","also_buy","also_view", "similar_item", "price"]], how="inner", on="itemId")
```

```python
data.drop_duplicates("itemId", inplace=True)
```

```python
data = data.iloc[:10000]
data
```

```python
data["title"]= data["title"].apply(lambda x:x.split())
data["title"] = data["title"].apply(lambda x:[i.replace(" ", " ") for i in x ])

data["brand"]= data["brand"].apply(lambda x:x.split())
data["brand"] = data["brand"].apply(lambda x:[i.replace(" ", " ") for i in x ])

data["main_cat"]= data["main_cat"].apply(lambda x:x.split())
data["main_cat"] = data["main_cat"].apply(lambda x:[i.replace(" ", " ") for i in x ])

data["similar_item"]= data["similar_item"].apply(lambda x:x.split())
data["similar_item"] = data["similar_item"].apply(lambda x:[i.replace(" ", " ") for i in x ])

data["price"]= data["price"].apply(lambda x:x.split())
data["price"] = data["price"].apply(lambda x:[i.replace(" ", " ") for i in x ])
```

```python
data
```

```python
data["tags"]= data["title"] + data["brand"] + data["category"] + data["main_cat"] + data["also_buy"] + data["also_view"] + data["price"]

data["tags"] = data["tags"].apply(lambda x :" ".join(x))
data["main_cat"] = data["main_cat"].apply(lambda x :" ".join(x))
data["brand"] = data["brand"].apply(lambda x :" ".join(x))
data["title"] = data["title"].apply(lambda x :" ".join(x))
data["tags"]= data["tags"].apply(lambda x:x.lower())
```

```python
data["tags"][5]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y= []
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
def stem(text):
    y= []
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
data["tags"]= data["tags"].apply(stem)
```

```python
data["tags"][5]
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

cv = TfidfVectorizer(lowercase=True, stop_words = "english")
vectors = cv.fit_transform(data["tags"]).toarray()
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similaritiy = cosine_similarity(vectors)
similaritiy
```

```python
def reccomend(movie):
    movie_index = data[data['title'] == movie].index[0]
    distances = similaritiy[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(data.iloc[i[0]].title)
        print("Similarity values: {}".format(i[1]))
        print(50*"-")
```

```python
reccomend("Netgear RT314 Internet Access Router")
```
