```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import pickle
```

```python
ratings= pickle.load(open("svd_ratings_data.pkl","rb" ))
```

```python
content_based_data = pickle.load(open("content_based_data.pkl","rb" ))
content_based_similaritiy = pickle.load(open("content_based_similaritiy.pkl","rb" ))
```

```python
ratings = ratings[ratings['rating'] == 5]
ratings = pd.merge(ratings,  content_based_data[["itemId","title","genres"]], how="inner", on="itemId")
```

```python

def reccomend(userıd): 
    user_movies = list(ratings[ratings["userId"]==userıd].sample(5).title) # I chose random 5 sample movies in that clsuter
    
    for movie in user_movies:
        movie_index = content_based_data[content_based_data['title'] == movie].index[0]
        distances = content_based_similaritiy[movie_index]
        movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:3] # I recommend 2 movies for per random sample movies
        
        for i in movies_list:
            print(content_based_data.iloc[i[0]].title, content_based_data.iloc[i[0]].genres)
```

```python
reccomend(15)
```
