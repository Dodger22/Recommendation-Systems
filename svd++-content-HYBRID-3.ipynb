```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import pickle
```

```python
content_based_data = pickle.load(open("content_based_data.pkl","rb" ))
content_based_similaritiy = pickle.load(open("content_based_similaritiy.pkl","rb" ))
ratings= pickle.load(open("svd_ratings_data.pkl","rb" ))
model = pickle.load(open("model_svd++.pkl","rb" ))
all_predictions = pickle.load(open("all_predictions_svd++.pkl","rb" ))
```

```python
all_predictions= pd.merge(all_predictions, content_based_data[["itemId", "title"]], how="inner", on="itemId")
ratings = pd.merge(ratings, content_based_data[["itemId","title","genres"]], how="inner", on="itemId")
```

```python
def reccomend(userıd):
    user_movies = all_predictions[all_predictions["userId"]==int(userıd)].sort_values(by='prediction', ascending=False)[:5].title
    
    for movie in user_movies:
        movie_index = content_based_data[content_based_data['title'] == movie].index[0]
        distances = content_based_similaritiy[movie_index]
        movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:3]
        
        for i in movies_list:
            print(content_based_data.iloc[i[0]].title, content_based_data.iloc[i[0]].genres)
```
